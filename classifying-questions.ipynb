```python
import numpy as np
import pandas as pd 
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from sklearn import preprocessing
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection, naive_bayes, svm
from sklearn.metrics import accuracy_score
import pickle
```

```python
df = pd.read_csv('/kaggle/input/labelleddddata/LabelledData (1) (1).txt',sep=',,,',engine='python',names=["Question", "Type"])
df
```

```python
x = df.iloc[:,:-1].values
x = pd.DataFrame(data=x)
y = df.iloc[:,1].values
```

```python
le = preprocessing.LabelEncoder()
le.fit_transform(y)
```

```python
corpus = []
for i in range(0, 1483):
    questions = x[0][i].lower()
    questions = questions.split()
    questions = [word for word in questions if word in set(stopwords.words('english'))]
    questions = ' '.join(questions)
    corpus.append(questions)
```

```python
np.random.seed(500)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(corpus, y, test_size = 0.20, random_state = 0)
```

```python
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(corpus)
Train_X_Tfidf = Tfidf_vect.transform(x_train)
Test_X_Tfidf = Tfidf_vect.transform(x_test)
```

```python
SVM = svm.SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
SVM.fit(Train_X_Tfidf,y_train)
predictions_SVM = SVM.predict(Test_X_Tfidf)
print("SVM Accuracy Score: ",accuracy_score(predictions_SVM, y_test)*100)
```

```python
filename = 'finalized_model.sav'
pickle.dump(SVM, open(filename, 'wb'))
```
